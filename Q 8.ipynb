```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.service import Service
```

```python
drivers = webdriver.Chrome(r"C:\Users\Dell\Videos\chromedriver_win32\chromedriver.exe")

drivers.get('https://www.forbes.com/billionaires/')
```

```python
# rank

Rank = []

rank = drivers.find_elements(By.XPATH, '//div[@class= "rank"]')

for i in rank:
    Rank.append(i.text)

```

```python
# Name

Name = []

name = drivers.find_elements(By.XPATH, '//div[@class="personName"]')

for i in name:
    Name.append(i.text)
    
```

```python
# Net Worth

Net_Worth = []

networth = drivers.find_elements(By.XPATH, '//div[@class="netWorth"]')
    
for i in networth:
    Net_Worth.append(i.text)
        



```

```python
# age 

Age = []

age = drivers.find_elements(By.XPATH, '//div[@class="age"]')

for i in age:
    Age.append(i.text)
```

```python
# citizen

Country = []

country = drivers.find_elements(By.XPATH, '//div[@class="countryOfCitizenship"]')

for i in country:
    Country.append(i.text)
```

```python
# source-text

Source = []

source = drivers.find_elements(By.XPATH, '//span[@class="source-text"]')

for i in source:
    Source.append(i.text)
```

```python
# category

Industry = []

industry = drivers.find_elements(By.XPATH, '//div[@class="category"]')

for i in industry:
    Industry.append(i.text)

```

```python
import pandas as pd

final = pd.DataFrame()

final['Rank']= Rank
final['Name']= Name
final['Net_Worth']= Net_Worth
final['Age']= Age
final['Country']= Country
final['Source']= Source
final['Industry']= Industry

final
```
