```python
# data 

Title1 = []

title = drivers.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal" or @class="a-size-base-plus a-color-base a-text-normal"]')

for i in title:
    Title1.append(i.text)
Title2=pd.Series(Title1).drop([30])
```

```python
Price = []

price = drivers.find_elements(By.XPATH, '//span[contains(@class, "a-price-whole")]')

for i in price:
    Price.append(i.text)

Price1=pd.Series(Price).drop([30,31,32,33,34,35,36,37])
```

```python
URL = []

url = drivers.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')

for i in url:
    
    URL.append(i.get_attribute('href'))
len(URL)

```

```python
Brand = []
for i in Title1:
    a=i.split()[0]
    Brand.append(a)
```

```python
Brand1=pd.Series(Brand).drop([30])
```

```python
Final =pd.DataFrame()

Final['Brand']= Brand1
Final['Product_Name']= Title2
Final['Price']= Price1
Final['URL']= URL

Final
```
