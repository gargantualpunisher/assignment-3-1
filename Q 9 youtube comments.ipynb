```python
drivers = webdriver.Chrome(r"C:\Users\Dell\Videos\chromedriver_win32\chromedriver.exe")

drivers.get('https://www.youtube.com/watch?v=WOdnRhWeHoY')
```

```python
# who commented
Time = []
time = drivers.find_elements(By.XPATH, '//*[@id="header-author"]/yt-formatted-string/a')

for i in time[0:500]:
    Time.append(i.text)
len(Time) 
```

```python
comments = drivers.find_elements(By.XPATH, '//*[@id="content-text"]')

Comments =[]

for i in comments[0:500]:
    Comments.append(i.text)
len(Comments)
```
