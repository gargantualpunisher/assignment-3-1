```python
# product

Name = []

name = drivers.find_elements(By.CLASS_NAME, "_4rR01T")

for i in name:
    Name.append(i.text)
len(Name)
```

```python
Colour=[]

for i in Name:
    e = slice(-21,-7)
    Colour.append(i[e])
len(Colour)
```

```python
Brand = []
for i in Name:
    a=i.split()[0]
    Brand.append(a)
Brand
```
