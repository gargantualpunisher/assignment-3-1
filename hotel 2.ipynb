```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.service import Service
```

```python
drivers = webdriver.Chrome(r"D:\selenium\chromedriver.exe")

drivers.get('https://www.hostelworld.com/')

```

```python
# CLICK
click2 = drivers.find_element(By.ID, "location-text-input-field")
click2.click()
```

```python
# input 

input = drivers.find_element(By.ID, "search-input-field")
input.send_keys('London')
```

```python
# click

click1=drivers.find_element(By.ID, "search-button")
click1.click()
```

```python
# Name

Name = []
name = drivers.find_elements(By.XPATH, '//h2[@class="title title-6"]')

for i in name:
    Name.append(i.text)
len(Name)
```

```python
# RATING
RATING =[]

rating = drivers.find_elements(By.XPATH, '//div[@class="score orange big" or @class= "score gray big"]')
for i in rating:
    RATING.append(i.text)


```

```python
# Price
Price =[]
price = drivers.find_elements(By.XPATH, '//div[@class="price-col"]')

for i in price:
    Price.append(i.text)
Price3 =pd.Series(Price)
Price3
```

```python
Price7 = []
Price9 = []

for i in Price3:
    if 'Privates' in i:
        Price7.append(i)
        
    else:
        Price9.append(i)
        
```
